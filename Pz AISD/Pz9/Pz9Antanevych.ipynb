**Пояснення:**
1. Алгоритм Гафмена дає **оптимальний префіксний код**
2. Найчастіший символ (A) отримав найкоротший код (1 біт)
3. Рідкісний символ (D) отримав довший код (3 біти)

| №  | Контрольнi питання                                                                 | Відповідь                                                                                     |
|----|-----------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------|
| 1  | Що таке кодування Гаффмена та як воно працює?                                    | Метод стиснення даних, що використовує змінну довжину кодів для символів (часті символи кодуються коротше). Працює через побудову двійкового дерева на основі частоти символів. |
| 2  | Як визначається оптимальний двійковий код Гаффмена для стиснення даних?         | <ul><li>Аналіз частоти символів у вхідних даних</li><li>Побудова дерева Гаффмена (часті символи ближче до кореня)</li><li>Генерація кодів на основі шляхів у дереві</li></ul> |
| 3  | Які переваги має кодування Гаффмена над іншими методами стиснення даних?        | <ul><li>Оптимальність (досягає межі ентропії Шеннона)</li><li>Адаптивність до частот символів</li><li>Відсутність втрат даних (безвтратне стиснення)</li></ul> |
| 4  | Як відбувається декодування даних, закодованих за допомогою кодування Гаффмена? | <ol><li>Використання того ж дерева Гаффмена, що й при кодуванні</li><li>Послідовне проходження по дереву за бітами коду</li><li>Повернення символу при досягненні листа дерева</li></ol> |
