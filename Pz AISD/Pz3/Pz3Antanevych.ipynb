## –í–∞—Ä—ñ–∞–Ω—Ç 1  

**–ó–∞–≤–¥–∞–Ω–Ω—è:** 

–ú–∞—î–º–æ –¥–≤—ñ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùëõ) = 4ùëõ3 ‚àí 12ùëõ + 9, ùëî(ùëõ) = ùëõ3. –î–æ–≤–µ—Å—Ç–∏, —â–æùëì(ùëõ) = Œ©(ùëî(ùëõ)).

### –†–æ–∑–≤'—è–∑–æ–∫: 

## –ê–ª–≥–æ—Ä–∏—Ç–º –±—É–ª—å–±–∞—à–∫–æ–≤–æ–≥–æ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
```
