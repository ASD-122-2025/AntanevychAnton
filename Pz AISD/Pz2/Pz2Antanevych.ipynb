## Варіант 1  
**Завдання:** 

Маємо дві функції 𝑓(𝑛) = 4𝑛3 − 12𝑛 + 9, 𝑔(𝑛) = 𝑛3. Довести, що𝑓(𝑛) = Ω(𝑔(𝑛)).

### Розв'язок: 

**Умова:**  
Дано функцію f(n) = 4n³ - 12n + 9. Потрібно довести, що вона зростає не повільніше за g(n) = n³.

### Розв'язок:
**Порівняння функцій**  
Поділимо обидві функції на n³:
f(n)/g(n) = (4n³ - 12n + 9)/n³ = 4 - 12/n² + 9/n³
**Аналіз поведінки**  
При великих значеннях n:
12/n² → 0
9/n³ → 0
Отже, f(n)/g(n) ≈ 4

**Перевірка нерівності**  
Виберемо множник 2 (менше за 4) і перевіримо:
4n³ - 12n + 9 ≥ 2n³
2n³ - 12n + 9 ≥ 0
**Конкретні значення**  
Для n=2:
2*8 - 12*2 + 9 = 16 - 24 + 9 = 1 ≥ 0
Для всіх n ≥ 2 нерівність виконується.

**Висновок:**  
Функція f(n) дійсно зростає не повільніше за n³, починаючи з n=2.

## Варіант 6 
**Завдання:** 

Нехай 𝑓(𝑛) = 6𝑛3 − 9𝑛2 + 12𝑛 і 𝑔(𝑛) = 𝑛3. Довести, що 𝑓(𝑛) =𝛺(𝑔(𝑛)).

### Розв'язок задачi: 
**Умова:**  
Дано функцію f(n) = 6n³ - 9n² + 12n. Потрібно довести, що вона зростає не повільніше за g(n) = n³.

### Розв'язок:
**Порівняння функцій**  
Поділимо обидві функції на n³:
f(n)/g(n) = 6 - 9/n + 12/n²
**Аналіз поведінки**  
При великих n:
9/n → 0
12/n² → 0
Отже, f(n)/g(n) ≈ 6

**Перевірка нерівності**  
Виберемо множник 3 (половина від 6):
6n³ - 9n² + 12n ≥ 3n³
3n³ - 9n² + 12n ≥ 0
**Факторизація**  
Винесемо 3n:
3n(n² - 3n + 4) ≥ 0
Квадратний вираз завжди додатній (D = 9-16 = -7 < 0).

**Висновок:**  
Функція f(n) зростає не повільніше за n³ для всіх n ≥ 1.
