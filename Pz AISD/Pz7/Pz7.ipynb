## Варіант 1  

Маємо дві короткі послідовності символів: «ABCDF» і «ACEDB».
Знайти найдовшу спільну підпослідовності символів, використовуючи алгоритм
динамічного програмування. 

### Розв'язок: 

 Знаходження найдовшої спільної підпослідовності (НСП)

## Вхідні дані
Послідовність X: `A B C D F`  
Послідовність Y: `A C E D B`

## Алгоритм динамічного програмування

### Крок 1: Ініціалізація матриці
Створимо матрицю `dp` розміром (m+1)×(n+1), де m=5, n=5

### Крок 3: Визначення НСП
1. Починаємо з dp[5][5] = 3
2. Рухаємось назад:
   - X[5]=F ≠ Y[5]=B → переходимо до max(dp[4][5], dp[5][4]) = 3
   - X[4]=D == Y[4]=D → додаємо 'D' до результату
   - Переходимо до dp[3][3]
   - X[3]=C ≠ Y[3]=E → max(dp[2][3], dp[3][2]) = 2
   - X[3]=C == Y[2]=C → додаємо 'C' до результату
   - Переходимо до dp[2][1]
   - X[2]=B ≠ Y[1]=A → max(dp[1][1], dp[2][0]) = 1
   - X[1]=A == Y[1]=A → додаємо 'A' до результату
