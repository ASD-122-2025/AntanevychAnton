### Розв'язок: 

# Топологічне сортування ациклічного графа за алгоритмом Кана

## Заданий граф
**Вершини:** {1, 2, 3, 4, 5, 6, 7}  
**Ребра:** {(1,2), (1,3), (1,6), (3,7), (1,4), (4,5)}

## Графічне представлення

<pre>
      1
    / | \ \
  2   3  4 6
     /    \
    7      5
</pre>

**Ініціалізація черги з вершинами, що мають вхідний степінь 0:**
   - Черга: [1]

**Процес сортування:**
   - Видаляємо 1, додаємо до результату: [1]
     - Оновлюємо степені сусідів:
       - 2: 0, 3: 0, 4: 0, 6: 0
     - Додаємо нові вершини з нульовим степенем до черги: [2, 3, 4, 6]
   - Видаляємо 2, додаємо до результату: [1, 2]
     - Немає сусідів для оновлення
   - Видаляємо 3, додаємо до результату: [1, 2, 3]
     - Оновлюємо степінь для 7: 0
     - Додаємо 7 до черги: [4, 6, 7]
   - Видаляємо 4, додаємо до результату: [1, 2, 3, 4]
     - Оновлюємо степінь для 5: 0
     - Додаємо 5 до черги: [6, 7, 5]
   - Видаляємо 6, додаємо до результату: [1, 2, 3, 4, 6]
     - Немає сусідів для оновлення
   - Видаляємо 7, додаємо до результату: [1, 2, 3, 4, 6, 7]
     - Немає сусідів для оновлення
   - Видаляємо 5, додаємо до результату: [1, 2, 3, 4, 6, 7, 5]

**Результат топологічного сортування:**
   - **1 → 2 → 3 → 4 → 6 → 7 → 5**
   - Інші можливі варіанти:
     - 1 → 2 → 3 → 7 → 4 → 5 → 6
     - 1 → 2 → 4 → 5 → 3 → 7 → 6
     - (Всього 6 можливих варіантів для даного графа)

# Валідація результату
Для кожного ребра (u → v) у відсортованому списку u завжди йде перед v, що задовольняє умовам топологічного порядку.

| №  | Контрольне питання                                                                 | Відповідь                                                                                     |
|----|-----------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------|
| 1  | Які переваги і недоліки алгоритму Кана порівняно з алгоритмом DFS для топологічного сортування графа? | <ul><li>**Алгоритм Кана**:<br>Простий у реалізації<br>Ефективний для розріджених графів<br>Вимагає попереднього обчислення вхідних степенів</li><li>**DFS-алгоритм**:<br>Гнучкіший<br>Може використовуватися для виявлення циклів<br>Рекурсія може призвести до переповнення стеку</li></ul> |
| 2  | Яка складність часу і пам’яті для кожного з алгоритмів у найгіршому і найкращому випадках? | <ul><li>**Алгоритм Кана**:<br>Час: O(V+E) у всіх випадках<br>Пам'ять: O(V) (для зберігання вхідних степенів)</li><li>**DFS-алгоритм**:<br>Час: O(V+E) у всіх випадках<br>Пам'ять: O(V) (для рекурсивного стеку)</li></ul> |
| 3  | Чи можна застосовувати алгоритм Кана до графів з вагами на ребрах? Як це порівняти з DFS? | <ul><li>**Алгоритм Кана**:<br>Може працювати з вагами, але не враховує їх при сортуванні<br>Ваги можна використовувати лише для додаткових обчислень</li><li>**DFS-алгоритм**:<br>Краще підходить для роботи з вагами<br>➖ Може враховувати ваги при побудові порядку</li></ul> |
| 4  | Як впливає структура графа на швидкість роботи кожного з цих алгоритмів? | <ul><li>**Алгоритм Кана**:<br>Краще працює на розріджених графах<br>Чутливий до кількості вхідних ребер</li><li>**DFS-алгоритм**:<br>Ефективніший на щільних графах<br>Менш залежний від структури</li></ul> |
| 5  | Чи є обмеження використання кожного алгоритму для певних типів графів або завдань? | <ul><li>**Алгоритм Кана**:<br>Не підходить для графів з циклами (вимагає попередньої перевірки)<br>Менш ефективний для графів зі складними залежностями</li><li>**DFS-алгоритм**:<br>Може мати проблеми з дуже великими графами через рекурсію<br>Вимагає додаткових перевірок для виявлення циклів</li></ul> |
| 6  | Які варіанти оптимізації можна застосувати для кожного алгоритму з метою поліпшення його продуктивності? | <ul><li>**Алгоритм Кана**:<br>Використання черги з пріоритетами<br>Паралельне обчислення вхідних степенів</li><li>**DFS-алгоритм**:<br>Ітеративна реалізація замість рекурсивної<br>Кешування проміжних результатів</li></ul> |
