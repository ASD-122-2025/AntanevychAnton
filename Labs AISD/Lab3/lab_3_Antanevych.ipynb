## Визначення асимптотичної складності алгоритмів сортування

- **Сортування вставками (Insertion Sort):** `O(n²)`
- **Сортування бульбашкою (Bubble Sort):** `O(n²)`

> Це означає, що зі збільшенням розміру вхідних даних (n) у 2 рази, час виконання зростає приблизно в 4 рази.

---

## Реалізація алгоритмів

**Реалізація на Python:**


```python
# Алгоритм сортування вставками:
def insertion_sort(nums):
    for j in range(1, len(nums)):
        key = nums[j]
        i = j - 1
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        nums[i + 1] = key

# Приклад
arr = [5, 3, 8, 1, 2]
insertion_sort(arr)
print(arr)        
```

```python
# Алгоритм сортування бульбашкою
def bubble_sort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

# Приклад
arr = [5, 3, 8, 1, 2]
bubble_sort(arr)
print(arr)
```

```python
# Вимірювання часу виконання
import time
def measure_time(sort_func, data):
    start_time = time.time()
    sort_func(data.copy())
    return time.time() - start_time
```

```python
# Графіки порівняння залежності часу роботи від розміру даних
%matplotlib inline
import matplotlib.pyplot as plt

# Розміри вхідних масивів
input_sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]

# Змодельовані дані часу виконання (секунди)
bubble_sort_time = [0.0001, 0.0002, 0.0008, 0.003, 0.013, 0.042, 0.11, 0.23, 0.45, 0.85, 2.0, 6.5, 17.9, 50.0]
insertion_sort_time = [0.00015, 0.00025, 0.0009, 0.0045, 0.018, 0.048, 0.095, 0.18, 0.32, 0.55, 1.1, 3.2, 9.8, 28.0]

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, bubble_sort_time, marker='D', linestyle='-', color='blue', label='Бульбашкове сортування')
plt.plot(input_sizes, insertion_sort_time, marker='^', linestyle='--', color='green', label='Сортування вставками')

plt.title("Залежність часу роботи алгоритмів від розміру даних")
plt.xlabel("Кількість елементів у масиві")
plt.ylabel("Час виконання (секунди)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Контрольні питання

1. **Що таке O-нотація і чим вона відрізняється від Ω-нотації?**  
   - **O-нотація (Big-O)** описує верхню межу зростання часу або ресурсів, які споживає алгоритм у найгіршому випадку.  
   - **Ω-нотація (Big-Omega)** вказує на нижню межу — тобто мінімальний час виконання в найкращому випадку.  
   - Вони відрізняються напрямком оцінки: `O()` — "не більше ніж", `Ω()` — "не менше ніж".

2. **Яку часову складність має алгоритм сортування методом вставляння за найсприятливіших умов, коли масив відсортовано? Відповідь обґрунтувати.**  
   - У найкращому випадку, коли масив вже відсортований, Insertion Sort має **лінійну складність `O(n)`**,  
     оскільки кожен елемент просто порівнюється з попереднім і не пересувається.

3. **Надати визначення ефективного алгоритму.**  
   - Ефективний алгоритм — це такий, що розв’язує задачу за розумний (поліноміальний) час і з мінімальними витратами ресурсів (пам’яті, енергії тощо),  
     особливо при зростанні розміру вхідних даних.

4. **Пояснити поняття головного параметра (розміру) завдання.**  
   - Головний параметр — це змінна, яка характеризує розмір вхідних даних.  
     Наприклад, для масиву це кількість елементів `n`, для графа — кількість вершин або ребер.

5. **Функція часової складності має вигляд: F(N) = N³ + 7N² − 14N. Як записати асимптотичну складність у O-нотації?**  
   - При великому `N` найважливішим є доданок з найвищим степенем — `N³`.  
     Отже, асимптотична складність: **`O(N³)`**.
