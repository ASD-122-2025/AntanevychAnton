1. Реалізацiя процедури обчислення факторіалу за допомогою циклу for та оцiнка асимптотичної складностi алгоритму.

```python
#Обчислення факторіалу за допомогою циклу for
def factorial_for(n):
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial
```

| Асимптотична складність | Пояснення                                                                                     |
|-------------------------|----------------------------------------------------------------------------------------------|
| O(n)                    | Алгоритм виконує n ітерацій циклу, кожна з яких має константну складність O(1).               |

2. Оцiнка асимптотичної складностi рекурсивного алгоритму обчислення n-го числа Фібоначчі.

```python
#Рекурсивний алгоритм обчислення факторіалу
def fac(n):
    if n == 0:
        return 1
    return fac(n - 1) * n
```

| Асимптотична складність | Пояснення                                                                                     |
|-------------------------|----------------------------------------------------------------------------------------------|
| O(n)                    | Алгоритм робить n рекурсивних викликів, кожен з яких має складність O(1).                     |

3. Оцiнка асимптотичної складностi рекурсивного алгоритму обчислення факторіалу.

```python
#Рекурсивний алгоритм обчислення чисел Фібоначчі
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
```

| Асимптотична складність | Пояснення                                                                                     |
|-------------------------|----------------------------------------------------------------------------------------------|
| O(2ⁿ)                   | Кожен виклик генерує 2 нових виклики, що призводить до експоненційного зростання.            |

4. Оцiнка асимптотичної складностi алгоритму сортування злиттям, використовуючи головну теорему рекурсії.

```python
#Алгоритм сортування злиттям (Merge Sort)
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(list):
    if len(list) < 2:
        return list
    middle = len(list) // 2
    left = mergesort(list[:middle])
    right = mergesort(list[middle:])
    return merge(left, right)
```

| Асимптотична складність | Пояснення                                                                                     |
|-------------------------|----------------------------------------------------------------------------------------------|
| O(n log n)              | За головною теоремою рекурсії: час розбиття списку навпіл (log n) та об'єднання (n).          |
