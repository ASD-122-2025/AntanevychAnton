# Завдання для самостійної роботи
### Завдання 1: Бінарні дерева
Створюэмо бінарне дерево з реалiзацiєю процедури видалення гiлки дерева.

```python
class BinaryTree:
    def __init__(self, root):
        self.root = root
        self.left_child = None
        self.right_child = None
    
    def insert_left(self, new_node):
        if self.left_child is None:
            self.left_child = BinaryTree(new_node)
        else:
            new_tree = BinaryTree(new_node)
            new_tree.left_child = self.left_child
            self.left_child = new_tree
    
    def insert_right(self, new_node):
        if self.right_child is None:
            self.right_child = BinaryTree(new_node)
        else:
            new_tree = BinaryTree(new_node)
            new_tree.right_child = self.right_child
            self.right_child = new_tree
    
    def get_left_child(self):
        return self.left_child
    
    def get_right_child(self):
        return self.right_child
    
    def set_root(self, value):
        self.root = value
    
    def get_root(self):
        return self.root
    
    def delete_branch(self, value):
        """Видалення гілки з заданим значенням"""
        if self.left_child and self.left_child.get_root() == value:
            self.left_child = None
            return True
        if self.right_child and self.right_child.get_root() == value:
            self.right_child = None
            return True
        
        deleted = False
        if self.left_child:
            deleted = self.left_child.delete_branch(value)
        if not deleted and self.right_child:
            deleted = self.right_child.delete_branch(value)
        return deleted
    
    def search(self, value):
        """Пошук значення у дереві"""
        if self.root == value:
            return True
        found = False
        if self.left_child:
            found = self.left_child.search(value)
        if not found and self.right_child:
            found = self.right_child.search(value)
        return found
    
    def __str__(self):
        return f"BinaryTree({self.root}, left={self.left_child}, right={self.right_child})"


# Приклад використання
if __name__ == "__main__":
    # 1. Створення дерева
    tree = BinaryTree('a')
    tree.insert_left('b')
    tree.insert_right('c')
    tree.get_left_child().insert_left('d')
    tree.get_left_child().insert_right('e')
    tree.get_right_child().insert_left('f')
    
    print("Початкове дерево:")
    print(tree)
    
    # 2. Видалення гілки
    print("\nВидаляємо гілку 'b':")
    tree.delete_branch('b')
    print(tree)
    
    # 3. Пошук елемента
    print("\nПошук елемента 'f':", tree.search('f'))
    print("Пошук елемента 'x':", tree.search('x'))
```

### Аналіз складності операцій:

1. **Пошук (search)**:
   - Середній випадок: O(n)
   - Найгірший випадок: O(n) (коли елемента немає або він останній)

2. **Вставка (insert)**:
   - Середній випадок: O(1) для конкретного місця вставки
   - Найгірший випадок: O(n) якщо потрібно знайти місце вставки

3. **Видалення (delete)**:
   - Середній випадок: O(n)
   - Найгірший випадок: O(n) (коли потрібно видалити листок або елемента немає)

Примітка: n - кількість вузлів у дереві.

### Завдання 2: Робота з купами
Функція генерації купи з випадкового масиву з видаленням максимального елементу з купи.

```python
import random

class MaxHeap:
    def __init__(self):
        self.heap = []
    
    def _parent(self, i):
        return (i - 1) // 2
    
    def _left_child(self, i):
        return 2 * i + 1
    
    def _right_child(self, i):
        return 2 * i + 2
    
    def _sift_up(self, i):
        while i > 0 and self.heap[self._parent(i)] < self.heap[i]:
            self.heap[self._parent(i)], self.heap[i] = self.heap[i], self.heap[self._parent(i)]
            i = self._parent(i)
    
    def _sift_down(self, i):
        max_index = i
        n = len(self.heap)
        
        l = self._left_child(i)
        if l < n and self.heap[l] > self.heap[max_index]:
            max_index = l
        
        r = self._right_child(i)
        if r < n and self.heap[r] > self.heap[max_index]:
            max_index = r
        
        if i != max_index:
            self.heap[i], self.heap[max_index] = self.heap[max_index], self.heap[i]
            self._sift_down(max_index)
    
    # 1. Генерація купи з випадкового масиву
    def build_heap(self, array):
        self.heap = array.copy()
        n = len(self.heap)
        for i in range(n // 2 - 1, -1, -1):
            self._sift_down(i)
    
    # 2. Додавання елемента (номеру з списку групи)
    def insert(self, value):
        self.heap.append(value)
        self._sift_up(len(self.heap) - 1)
    
    # 3. Видалення максимального елемента
    def extract_max(self):
        if len(self.heap) == 0:
            return None
        max_val = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self._sift_down(0)
        return max_val
    
    def search(self, value):
        return value in self.heap
    
    def __str__(self):
        return str(self.heap)

# Приклад використання
if __name__ == "__main__":
    heap = MaxHeap()
    
    # 1. Генерація купи з випадкового масиву
    random_array = [random.randint(1, 100) for _ in range(10)]
    print("Випадковий масив:", random_array)
    heap.build_heap(random_array)
    print("Купа після побудови:", heap)
    
    # 2. Додавання номера групи (номер 1)
    group_number = 1
    heap.insert(group_number)
    print(f"\nКупа після додавання номера групи ({group_number}):", heap)
    
    # 3. Видалення максимального елемента
    max_val = heap.extract_max()
    print(f"\nВидалено максимальний елемент: {max_val}")
    print("Купа після видалення максимуму:", heap)
    
    # 4. Пошук елемента
    search_value = group_number
    print(f"\nПошук значення {search_value}: {'знайдено' if heap.search(search_value) else 'не знайдено'}")
```

### Аналіз складності операцій:

1. **Побудова купи (build_heap)**:
   - Складність: O(n)

2. **Вставка (insert)**:
   - Середній випадок: O(1)
   - Найгірший випадок: O(log n)

3. **Видалення максимуму (extract_max)**:
   - Складність: O(log n)

4. **Пошук (search)**:
   - Складність: O(n) (лінійний пошук)

Примітка: n - кількість елементів у купі.

### Завдання 3: Геш-таблиці
Функцiя геш-таблицi з методом ланцюжків

```python
import time
import random
from string import ascii_letters

class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]
    
    def _hash(self, key):
        return hash(key) % self.size
    
    # Методи вирішення колізій (ланцюжками)
    def insert(self, key, value):
        index = self._hash(key)
        for item in self.table[index]:
            if item[0] == key:
                item[1] = value  # Оновлення значення, якщо ключ вже існує
                return
        self.table[index].append([key, value])
    
    def search(self, key):
        index = self._hash(key)
        for item in self.table[index]:
            if item[0] == key:
                return item[1]
        return None
    
    def delete(self, key):
        index = self._hash(key)
        for i, item in enumerate(self.table[index]):
            if item[0] == key:
                del self.table[index][i]
                return True
        return False
    
    def __str__(self):
        return str(self.table)

def generate_random_string(length=5):
    return ''.join(random.choice(ascii_letters) for _ in range(length))

def test_hash_table(data_type, test_cases=1000):
    print(f"\nТестування для типу даних: {data_type.__name__}")
    ht = HashTable(size=100)
    times = {'insert': [], 'search': [], 'delete': []}
    
    # Генерація тестових даних
    if data_type == int:
        test_data = [random.randint(1, 100000) for _ in range(test_cases)]
    elif data_type == float:
        test_data = [random.uniform(1, 100) for _ in range(test_cases)]
    elif data_type == str:
        test_data = [generate_random_string() for _ in range(test_cases)]
    elif data_type == tuple:
        test_data = [(random.randint(1, 100), generate_random_string(3)) for _ in range(test_cases)]
    else:
        test_data = [random.randint(1, 100) for _ in range(test_cases)]
    
    # Тестування вставки
    start = time.time()
    for i, key in enumerate(test_data):
        ht.insert(key, f"value_{i}")
    times['insert'].append(time.time() - start)
    
    # Тестування пошуку
    start = time.time()
    for key in test_data:
        ht.search(key)
    times['search'].append(time.time() - start)
    
    # Тестування видалення
    start = time.time()
    for key in test_data:
        ht.delete(key)
    times['delete'].append(time.time() - start)
    
    # Виведення результатів
    print(f"Час вставки {test_cases} елементів: {times['insert'][0]:.6f} сек")
    print(f"Час пошуку {test_cases} елементів: {times['search'][0]:.6f} сек")
    print(f"Час видалення {test_cases} елементів: {times['delete'][0]:.6f} сек")
    
    return times

if __name__ == "__main__":
    # Тестування з різними типами даних
    data_types = [int, float, str, tuple]
    results = {}
    
    for dt in data_types:
        results[dt.__name__] = test_hash_table(dt, test_cases=1000)
    
    # Порівняння результатів
    print("\nПорівняльна таблиця:")
    print("{:<10} {:<15} {:<15} {:<15}".format(
        "Тип даних", "Час вставки", "Час пошуку", "Час видалення"))
    for dt in data_types:
        name = dt.__name__
        print("{:<10} {:<15.6f} {:<15.6f} {:<15.6f}".format(
            name,
            results[name]['insert'][0],
            results[name]['search'][0],
            results[name]['delete'][0]))
```

### Аналіз результатів та висновки:

1. **Методи вирішення колізій**:
   - Використано метод ланцюжків (chaining)
   - Кожен слот таблиці містить список пар (ключ, значення)
   - При колізії нові елементи додаються у відповідний список

2. **Складність операцій**:
   - Вставка (insert): 
     - В середньому O(1), у найгіршому випадку O(n)
   - Пошук (search):
     - В середньому O(1), у найгіршому випадку O(n)
   - Видалення (delete):
     - В середньому O(1), у найгіршому випадку O(n)

3. **Порівняння типів даних**:
   - Найшвидші операції з числовими типами (int, float)
   - Операції з рядками та кортежами трохи повільніші через складність хешування

4. **Оптимізація**:
   - Для великих даних варто збільшувати розмір таблиці
