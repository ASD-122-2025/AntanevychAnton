**Завдання для самостійної роботи:**  

1. Оцінка асимптотичної складності алгоритмів

| Алгоритм         | Складність  | Пояснення                                                                 |
|------------------|-------------|--------------------------------------------------------------------------|
| Лінійний пошук   | O(n)        | Перевіряє всі елементи послідовно у найгіршому випадку                  |
| Бінарний пошук   | O(log n)    | На кожному кроці ділить масив навпіл, швидко звужуючи область пошуку    |

2. Функція для лінійного пошуку

```python
import time
import matplotlib.pyplot as plt

# 1. Визначаємо функції пошуку
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

3. Функція для бінарного пошуку з вимірюванням часу

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

4. Функція для вимірювання часу

```python
def measure_time(search_func, arr_size, search_func_name):
    arr = list(range(arr_size))
    target = arr_size - 1  # Шукаємо останній елемент
    
    start_time = time.time()
    search_func(arr, target)
    return time.time() - start_time
    
    # 3. Параметри для тестування
sizes = list(range(10, 10001, 500))  # Розміри масивів від 10 до 10000 з кроком 500
linear_times = []
binary_times = []
```

4. Графiк результатiв результатів

```python
for size in sizes:
    linear_times.append(measure_time(linear_search, size, "linear"))
    binary_times.append(measure_time(binary_search, size, "binary"))

# 5. Будуємо графік
plt.figure(figsize=(12, 6))
plt.plot(sizes, linear_times, 'b-', label='Лінійний пошук (O(n))')
plt.plot(sizes, binary_times, 'r-', label='Бінарний пошук (O(log n))')
plt.xlabel('Розмір масиву')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння продуктивності алгоритмів пошуку')
plt.legend()
plt.grid(True)
plt.show()
```

5. Висновки про ефективність алгоритмів

- **Бінарний пошук** демонструє значно кращу продуктивність для масивів розміром > 100 елементів
- **Лінійний пошук** може бути ефективнішим для дуже малих масивів (n < 10) через відсутність накладних витрат
- Для відсортованих даних бінарний пошук завжди є оптимальним вибором
- При необхідності частого пошуку варто використовувати бінарний пошук після попереднього сортування даних

> **Оптимальний вибір алгоритму** залежить від:
> 1. Частоти пошукових операцій
> 2. Необхідності підтримки відсортованого стану даних
> 3. Очікуваного розміру вхідних даних
